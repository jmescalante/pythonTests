```python
# We import all of the modules we'll use
%pylab inline
rcParams['figure.figsize'] = (10, 4) #wide graphs by default
from __future__ import print_function
from __future__ import division
```

```python
# we import the main image
img = imread('CORRELATION_Bach_original.png')
imshow(img)
```

```python
# We check the type
img.dtype
```

```python
# Now, I am not entirely sure what does this line does, clearly
# It has to do with the white/bright levels
img = sum(img.astype(float), axis=2)/4.0
imshow(img)
colorbar()
```

```python
#So, this are the dimensions of our image
img.shape
```

```python
# WE will turn the image to a grayscale one
imshow(img, cmap=cm.gray)
colorbar()
```

```python
# Here we are plotting the white values, are more evident
img = where(img > 0.1, -1, 1)
imshow(img, cmap='gray')
colorbar()
```

```python
# Now we will import the 'other image' (Bach eye)
ojo = imread('CORRELATION_Bach_eyeB.png')
imshow(ojo)
colorbar() 
```

```python
# We will do some threshold
ojo = ojo.astype(float).sum(axis=-1)/3
ojo = where(ojo>0.2,-1,1)
imshow(ojo, cmap=cm.gray, interpolation='nearest')
```

```python
# ANd now, finally, we'll do the correlate functionn
# We import the module first
from scipy.signal import correlate2d
```

```python
cc = correlate2d(img, ojo, mode ='same')
imshow(cc)
colorbar() # We shoe the color bar to get some real data
# gfc().set_figheight(8) # This does not work, why Lord? WHY!!?
# 
```

```python
imshow(where(cc > 1500, 1, 0), interpolation='nearest', cmap=cm.gray)
colorbar()
```

```python
from scipy.ndimage.filters import maximum_filter
imshow(maximum_filter(cc, (10,10)))
# NOW THIS IS VERY INTERESTING!!!! Could we generate sound from this "sources"
# More like a radar, this could give an idea for a whole different project (in the arts of course)
```

```python
# Let's see if there are any coincidences, and of course there are!!!!!!
subplot(121)
imshow(maximum_filter(cc, (50,50)))
subplot(122)
imshow(img, cmap=cm.gray)
```
