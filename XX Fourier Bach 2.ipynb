```python
# Now, let's make a function that scales a given value
# (pretty much like the map function in Processing)
# Could not find a similar in iPython, meh.

# IMportant, values must be given properly (max is greater than max)

def escalar( num, min1, max1, min2, max2 ):
    # PART A >> The previous place
    # We see which interval we are dealing with
    intervalo1 = max1 - min1
    # We re-arange the value to the interval
    position1 = num - min1
    # We calculate the percentage
    porcentaje1 = position1*100/intervalo1
    
    # PART B >> The new place
    # Now we will calculate the second interval
    intervalo2 = max2-min2
    # WE will calculate how much we should add
    adding2 = porcentaje1*intervalo2/100
    # We proceed to add the final number
    numeroFinal = min2+adding2
    
    # We return at last the number
    return numeroFinal
    
# numero = escalar(3, 0, 10, 0, 50)
# print(numero)
    
```
